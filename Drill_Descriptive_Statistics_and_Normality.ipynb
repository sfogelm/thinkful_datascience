```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta_dist = np.random.beta(a=2, b=2, size=100)
plt.hist(beta_dist)
plt.axvline(beta_dist.mean(), color='green', linestyle='solid', linewidth=2)
plt.axvline(beta_dist.mean() + beta_dist.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.axvline(beta_dist.mean() - beta_dist.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.show()
```

```python
binomial_dist = np.random.binomial(n=10, p=0.5, size=100)
plt.hist(binomial_dist)
plt.axvline(binomial_dist.mean(), color='green', linestyle='solid', linewidth=2)
plt.axvline(binomial_dist.mean() + binomial_dist.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.axvline(binomial_dist.mean() - binomial_dist.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.show()
```

```python
chisquare_dist = np.random.chisquare(df = 10, size=100)
plt.hist(chisquare_dist)
plt.axvline(chisquare_dist.mean(), color='green', linestyle='solid', linewidth=2)
plt.axvline(chisquare_dist.mean() + chisquare_dist.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.axvline(chisquare_dist.mean() - chisquare_dist.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.show()
```

```python
dirichlet_dist = np.random.dirichlet((10, 5, 3), 100)
plt.hist(dirichlet_dist)
plt.axvline(dirichlet_dist.mean(), color='green', linestyle='solid', linewidth=2)
plt.axvline(dirichlet_dist.mean() + dirichlet_dist.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.axvline(dirichlet_dist.mean() - dirichlet_dist.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.show()
```

```python
f_dist = np.random.f(dfnum=1, dfden=48, size=100)
plt.hist(f_dist)
plt.axvline(f_dist.mean(), color='green', linestyle='solid', linewidth=2)
plt.axvline(f_dist.mean() + f_dist.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.axvline(f_dist.mean() - f_dist.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.show()
```

```python
gamma_dist = np.random.gamma(shape=2, scale=2, size=100)
plt.hist(gamma_dist)
plt.axvline(gamma_dist.mean(), color='green', linestyle='solid', linewidth=2)
plt.axvline(gamma_dist.mean() + gamma_dist.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.axvline(gamma_dist.mean() - gamma_dist.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.show()
```

```python
var1= np.random.normal(5, 0.5, 1000)
plt.hist(var1)
plt.show()
```

```python
var2= np.random.normal(10,1,1000)
plt.hist(var2)
plt.show()
```

```python
var3 = var1 + var2
plt.hist(var3)
plt.axvline(var3.mean(), color='green', linestyle='solid', linewidth=2)
plt.axvline(var3.mean() + var3.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.axvline(var3.mean() - var3.std(), color='yellow', linestyle='dashed', linewidth=2)
plt.show()
```
